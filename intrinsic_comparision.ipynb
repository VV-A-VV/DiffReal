```python
import cv2
import matplotlib.pyplot as plt
image1_alb = cv2.imread('bedroom/breakfast-lamps_pred_alb.png')
image2_shd = cv2.imread('bedroom/breakfast-lamps_pred_shd.png')

# OpenCV  BGR
# corrected_image = cv2.cvtColor(image1_alb, cv2.COLOR_RGB2BGR)
# corrected_image = cv2.cvtColor(image2_shd, cv2.COLOR_RGB2BGR)

plt.figure(figsize=(20, 20))
plt.subplot(121)
plt.imshow(image1_alb)
plt.subplot(122)
plt.imshow(image2_shd)
```

```python
image1_alb = image1_alb / 255.0
image2_shd = image2_shd / 255.0
reconstructed_image = image1_alb * image2_shd
plt.figure(figsize=(20, 20))
plt.subplot(121)
plt.imshow(reconstructed_image)
plt.subplot(122)
plt.imshow(image1_alb)
plt.show()

```
